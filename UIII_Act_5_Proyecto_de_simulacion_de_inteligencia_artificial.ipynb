```python
# ==================================================
# SISTEMA INTELIGENTE PARA JOYERÍA
# CBTIS 128 - Inteligencia Artificial
# Mario Yañez
# Lista: 58
# Submódulo: Elabora Proyectos con Programación Lógica
# Docente: Ing. Eliseo Nava
# ==================================================

print("======================================")
print("   SISTEMA INTELIGENTE DE JOYERÍA")
print("======================================")

cliente = input("Nombre del cliente: ")
edad = int(input("Edad: "))
presupuesto = float(input("Presupuesto disponible: $"))
ocasion = input("Ocasión (compromiso, boda, aniversario, regalo): ").lower()

print("\nAnalizando información del cliente...")
print("--------------------------------------")

# Inteligencia Artificial simulada con IF, ELIF y ELSE

if ocasion == "compromiso" and presupuesto >= 20000:
    print("Recomendación IA:")
    print("💍 Anillo de diamante premium.")

elif ocasion == "compromiso" and presupuesto >= 10000:
    print("Recomendación IA:")
    print("💍 Anillo de oro blanco con zirconias.")

elif ocasion == "compromiso" and presupuesto < 10000:
    print("Recomendación IA:")
    print("💍 Anillo de plata con piedra brillante.")

elif ocasion == "boda" and presupuesto >= 25000:
    print("Recomendación IA:")
    print("💎 Juego de argollas de oro de 18 quilates.")

elif ocasion == "boda" and presupuesto < 25000:
    print("Recomendación IA:")
    print("💎 Juego de argollas de plata fina.")

elif ocasion == "aniversario" and presupuesto >= 15000:
    print("Recomendación IA:")
    print("📿 Collar de oro con piedras preciosas.")

elif ocasion == "aniversario" and presupuesto < 15000:
    print("Recomendación IA:")
    print("📿 Pulsera elegante de plata.")

elif ocasion == "regalo" and edad >= 18 and presupuesto >= 8000:
    print("Recomendación IA:")
    print("⌚ Reloj de lujo con grabado personalizado.")

elif ocasion == "regalo" and edad >= 18 and presupuesto < 8000:
    print("Recomendación IA:")
    print("🔗 Cadena de plata personalizada.")

elif ocasion == "regalo" and edad < 18:
    print("Recomendación IA:")
    print("✨ Pulsera juvenil moderna.")

else:
    print("No se encontró una recomendación específica.")
    print("Se recomienda una asesoría personalizada.")

print("--------------------------------------")
print("Cliente:", cliente)

if presupuesto >= 20000:
    print("Categoría del cliente: VIP")
elif presupuesto >= 10000:
    print("Categoría del cliente: Preferente")
else:
    print("Categoría del cliente: General")

print("--------------------------------------")
print("Gracias por utilizar el Sistema Inteligente de Joyería.")
print("CBTIS 128 - Inteligencia Artificial")
```
